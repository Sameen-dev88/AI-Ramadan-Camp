```python
# calculator App
Input1 = int(input("Enter value 1:"))
Input2 = int(input("Enter value 2:"))
operand = input("Enter your operand")
if operand == "+" :
    print(Input1 + Input2)
elif operand == "-" :
    print(Input1 - Input2)
elif operand == "*" :
    print(Input1 * Input2)
elif operand == "/" :
    print(Input1 / Input2)
else :
    print("Please enter correct operator")
```

```python
# question no 7
Input1 = int(input("Enter Maths Marks:"))
Input2 = int(input("Enter Science Marks:"))
Input3 = int(input("Enter Computer Marks:"))
Input4 = int(input("Enter Urdu Marks:"))
Input5 = int(input("Enter English Marks:"))
Obtain = Input1 + Input2 + Input3 + Input4 + Input5 
print("Your Obtain Marks : ",Obtain )
perc = Obtain / 1000 * 100
print("Your percentange is : ",perc )
if perc >= 80 and perc < 100 :
    print("A 1 Grade")
elif perc >= 70 and perc < 80 :
    print("A Grade")
elif perc >= 60 and perc < 70 :
        print("B Grade")
elif perc >= 50 and perc < 60 :
        print("C Grade")
elif perc >= 40 and perc < 50 :
        print("D Grade")
elif perc >= 33 and perc < 40 :
        print("E Grade")
else :
    print("Fail")
```
